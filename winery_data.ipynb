```python
import pandas as pd
```

```python
# import winery csv

file = 'Winery.csv'
```

```python
# check to see if winery imported correctly 

winery_df = pd.read_csv(file,)

winery_df.head()
```

```python
# delete columns that aren't neccessary to display

del winery_df['taster_name'] 
del winery_df['taster_twitter_handle']
del winery_df['points']
del winery_df['price']
del winery_df['variety']
del winery_df['description']

winery_df.head()
```

```python
# show only wineries from the US 

winery_df = winery_df.loc[winery_df['country'] == 'US']

winery_df.head()
```

```python
winery_df['country'].value_counts()
```

```python
# rename province column

winery_df.rename(columns={'province': 'state'}, inplace=True)

winery_df.head(10)
```

```python
# delete country column since we have a stat column 

del winery_df['country']

winery_df.head()
```

```python
# reorder columns to show title first

columns_names = ["title", "state", "region_1", "region_2", "winery", "designation"]
winery_df_clean = winery_df.reindex(columns = columns_names)

winery_df_clean.head()
```
