```python
import pandas as pd
import numpy as np
```

```python
#file to connect to
beers_file = "beers.csv"

brewery_file = "breweries.csv"
```

```python
#creating a pandas data frame with the craft beers csv file

craft_beers_df = pd.read_csv(beers_file, skiprows=[0])
craft_beers_df.head(100)
```

```python
craft_beers_df.info()
```

```python
#rename columns
cbeers_df_tidy = craft_beers_df.rename(columns = {'NaN': "Beer Number", 'name': "Craft Beer", "abv": 'ABV', "style": "Style", "brewery_id": "Brewery ID", "ounces": "Ounces"}, inplace = False)
cbeers_df_tidy.head()
```

```python
#reorder the columns

columns_names = ["Brewery ID", "Craft Beer", "Style", "ABV", "Ounces"]

cbeers_df_tidy2 = cbeers_df_tidy.reindex(columns = columns_names)

cbeers_df_tidy2
```

```python
#rename data frame
cbeers_df_final = cbeers_df_tidy2
```

```python
#brewery csv
#creating a pandas data frame with the craft beers csv file

breweries_df = pd.read_csv(brewery_file, skiprows=[0])
breweries_df.head(100)
```

```python
breweries_df.info
```

```python
#rename certain columns

breweries_df_tidy = breweries_df.rename(columns = {"Unnamed: 0": "Brewery ID", "name": "Brewery", "city": "City", "state": "State"}, inplace = False)
breweries_df_tidy
```

```python
#dataframe rename
brewery_df_final = breweries_df_tidy
```

```python
#merging the dataframes together through the Brewery ID
combined_df = pd.merge(cbeers_df_final, brewery_df_final, on='Brewery ID', how='inner')
combined_df
```

```python
# reorder combined data frame

Combined_column_names = ["Brewery ID", "Brewery", "Craft Beer", "Style", "ABV", "Ounces", "State", "City"]

combined_df_tidy = combined_df.reindex(columns = Combined_column_names)

combined_df_tidy
```

```python
#reorder rows based on Brewery ID

combined_df_tidy_final = combined_df_tidy.sort_values("Brewery ID")
combined_df_tidy_final
```
