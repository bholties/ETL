```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# import winery csv

file = 'resources\Winery.csv'
```

```python
# check to see if winery imported correctly 

winery_df = pd.read_csv(file,)

winery_df.head()
```

```python
# delete columns that aren't neccessary to display

del winery_df['taster_name'] 
del winery_df['taster_twitter_handle']
del winery_df['points']
del winery_df['price']
del winery_df['variety']
del winery_df['description']

winery_df.head()
```

```python
# show only wineries from the US 

winery_df = winery_df.loc[winery_df['country'] == 'US']

winery_df.head()
```

```python
winery_df['country'].value_counts()
```

```python
# rename province column

winery_df.rename(columns={'province': 'state'}, inplace=True)

winery_df.head(10)
```

```python
# delete country column since we have a stat column 

del winery_df['country']

winery_df.head()
```

```python
# reorder columns to show title first

columns_names = ["title", "state", "region_1", "region_2", "winery", "designation"]
winery_df_clean = winery_df.reindex(columns = columns_names)

winery_df_clean.head()
```

```python
# check number of blank values in columns region 1, region 2, and designation

winery_df_clean['region_1'].isnull().sum()
winery_df_clean['region_2'].isnull().sum()
winery_df_clean['designation'].isnull().sum()
```

```python
# delete rows with NaN (blank values)

winery_df_cleanup = winery_df_clean.dropna(subset = ['region_1'])
winery_df_cleanup2 = winery_df_cleanup.dropna(subset = ['region_2'])
winery_df_cleanup3 = winery_df_cleanup2.dropna(subset = ['designation'])

winery_df_cleanup3.head(10)
```

```python
#organize states by alphabetical order

final_winery = winery_df_cleanup3.sort_values(by='state',ascending = True)

final_winery.head(5).append(final_winery.tail(5))
```
