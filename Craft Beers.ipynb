```python
import pandas as pd
```

```python
#file to connect to
beers_file = "beers.csv"
```

```python
#creating a pandas data frame with the craft beers csv file

craft_beers_df = pd.read_csv(beers_file, skiprows=[0])
craft_beers_df.head(100)
```

```python
craft_beers_df.info()
```

```python
#delete unnecessary columns
del craft_beers_df['ibu']

del craft_beers_df['Unnamed: 0']

del craft_beers_df['id']
```

```python
#rename columns
cbeers_df_tidy = craft_beers_df.rename(columns = {'name': "Craft Beer", "abv": 'ABV', "style": "Style", "brewery_id": "Brewery ID", "ounces": "Ounces"}, inplace = False)
cbeers_df_tidy.head()
```

```python
#reorder the columns

columns_names = ["Brewery ID", "Craft Beer", "Style", "ABV", "Ounces"]

cbeers_df_tidy2 = cbeers_df_tidy.reindex(columns = columns_names)

cbeers_df_tidy2
```

```python
#rename data frame
cbeers_df_final = cbeers_df_tidy2
```
