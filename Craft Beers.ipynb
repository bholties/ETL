```python
#creating a pandas data frame with the craft beers csv file

craft_beers_df = pd.read_csv(beers_file, skiprows=[0])
craft_beers_df.head(100)
```

```python
craft_beers_df.info()
```

```python
#delete unnecessary columns
del craft_beers_df['ibu']

del craft_beers_df['Unnamed: 0']

del craft_beers_df['id']
```

```python
#rename columns
cbeers_df_tidy = craft_beers_df.rename(columns = {'name': "craft_beer"}, inplace = False)
cbeers_df_tidy.head()
```

```python
#reorder the columns

columns_names = ["brewery_id", "craft_beer", "style", "abv", "ounces"]

cbeers_df_tidy2 = cbeers_df_tidy.reindex(columns = columns_names)

cbeers_df_tidy2
```

```python
#rename data frame

cbeers_df_final = cbeers_df_tidy2
```

```python
cbeers_df_final['craft_beer'].value_counts()
```

```python
cbeers_df_final.drop_duplicates(subset=["brewery_id", "craft_beer"], keep='first', inplace=True)
```

```python
cbeers_df_final.to_sql("craft_beer", engine, if_exists = 'append', index=False)
```
